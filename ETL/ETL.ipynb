# ETL (Extracción, Transformación y Carga de datos)

##### Importar librerías necesarias 

```python
import pandas as pd
import numpy as np
```

##### Tenemos 2 archivos de excel 'homicidios' y 'lesiones' los cuales contiene las pestañas 'HECHOS' y 'VICTIMAS' con los que trabajaremos.

```python
h_hechos = pd.read_excel('../data/raw/homicidios.xlsx', sheet_name='HECHOS')
h_victimas = pd.read_excel('../data/raw/homicidios.xlsx', sheet_name='VICTIMAS')
l_hechos = pd.read_excel('../data/raw/lesiones.xlsx',sheet_name='HECHOS')
l_victimas = pd.read_excel('../data/raw/lesiones.xlsx',sheet_name='VICTIMAS')
```

### Dataframe h_hechos

```python
# Inspección preliminar de los datos

h_hechos.head()
```

##### Convertimos los nombres de las columnas a minúsculas

```python
h_hechos.columns = h_hechos.columns.str.lower()
```

##### Cambiamos el nombre de las columnas 'pos x' y 'pos y' por longitud y latitud respectivamente

```python
h_hechos.rename(columns={'pos x':'longitud', 'pos y':'latitud'})
```

### Tratamiento de Dataframe hom_victimas

```python
# Inspección preliminar de los datos

h_victimas.head()
```

##### Convertimos los nombres de las columnas a minúsculas

```python
h_victimas.columns = h_victimas.columns.str.lower()
```

##### Cambiamos el nombre de 'id_hecho' a 'id' para hacer coincidir los datos entre dataframes

```python
h_victimas.rename(columns={'id_hecho':'id'},inplace=True)
```

```python
# Verifiquemos el cambio.

h_victimas.head()
```

### Dataframe l_hechos

```python
# Inspección preliminar de los datos
l_hechos.head()
```

##### El diccionario de datos nos ayuda a establecer que los valores nulos de la columna 'gravedad' pueden ser considerados de gravedad leve. El resto de columnas no tiene alguna indicación que nos ayude  realizar algun cambio.

```python
l_hechos['gravedad'] = l_hechos['gravedad'].fillna('leve')
```

##### Renombramos columnas para mantener coincidencia entre columnas 

```python
l_hechos.rename(columns={'latutid':'latitud','franja_hora':'hh'},inplace=True)
```

### Dataframe les_victimas

```python
# Inspección preliminar de los datos

l_victimas.head()
```

##### El diccionario de datos nos ayuda a establecer que los valores nulos de la columna 'gravedad' pueden ser considerados de gravedad leve. El resto de columnas no tiene alguna indicación que nos ayude  realizar algun cambio.

```python
l_victimas['GRAVEDAD'] = l_victimas['GRAVEDAD'].fillna('leve')
```

##### Convertimos los nombres de las columnas a minúsculas

```python
l_victimas.columns = l_victimas.columns.str.lower()
```

```python
# Renombramos las columnas para que sean consistentes.

l_victimas.rename(columns={'id hecho':'id','aaa':'aaaa'})
```

### Concatenación de Dataframes


##### Un vez que tenemos el proceso de ETL realizado en cada dataframe, realizamos una concatenación entre los archivos similares para trabajar de forma más eficiente.

```python
# Concatenación de archivos por la pestaña 'HECHOS'

hechos = pd.concat([h_hechos, l_hechos], ignore_index=True)

# Concatenación de archivos por la pestaña 'VICTIMAS'

victimas = pd.concat([h_victimas, l_victimas], ignore_index=True)
```

#### Exportamos los Dataframes a archivos csv

```python
hechos.to_csv('../data/clean/hechos.csv', index=False)

victimas.to_csv('../data/clean/victimas.csv', index=False)
```
